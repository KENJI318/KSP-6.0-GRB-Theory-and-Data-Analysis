```python
import matplotlib.pyplot as plt
import numpy as np

from threeML import *
from threeML.io.package_data import get_path_of_data_file

from astropy.io.votable import parse
import pandas as pd
```

```python
votable = parse("votable.xml")
table = votable.get_first_table().to_table()

gbm_catalog = table.to_pandas()
df = gbm_catalog
```

```python
grb_info = gbm_catalog[gbm_catalog['name'] == "GRB080916009"]

detector_order = ["n0","n1","n2","n3","n4","n5","n6","n7","n8","n9","na","nb","b0","b1"]
mask_str = grb_info['scat_detector_mask'].values[0]
gbm_detectors = [det for det, bit in zip(detector_order, mask_str) if bit == '1']

source_interval = (
    float(grb_info['flnc_spectrum_start'].values[0]),
    float(grb_info['flnc_spectrum_stop'].values[0])
)
background_interval = [
    (float(grb_info['back_interval_low_start'].values[0]), float(grb_info['back_interval_low_stop'].values[0])),
    (float(grb_info['back_interval_high_start'].values[0]), float(grb_info['back_interval_high_stop'].values[0]))
]
model = grb_info['flnc_best_fitting_model'].values[0]
print(gbm_detectors)
```

```python
n3_tte = "Detectors/n3/glg_tte_n3_bn080916009_v01.fit"
n3_rsp = "Detectors/n3/glg_cspec_n3_bn080916009_v07.rsp"
n3_cspec = "Detectors/n3/glg_cspec_n3_bn080916009_v01.pha"

n4_tte = "Detectors/n4/glg_tte_n4_bn080916009_v01.fit"
n4_rsp = "Detectors/n4/glg_cspec_n4_bn080916009_v07.rsp"
n4_cspec = "Detectors/n4/glg_cspec_n4_bn080916009_v01.pha"

b0_tte = "Detectors/b0/glg_tte_b0_bn080916009_v01.fit"
b0_rsp = "Detectors/b0/glg_cspec_b0_bn080916009_v07.rsp"
b0_cspec = "Detectors/b0/glg_cspec_b0_bn080916009_v01.pha"

dload = {
    'n3': {'tte': n3_tte, 'rsp': n3_rsp, 'cspec': n3_cspec},
    'n4': {'tte': n4_tte, 'rsp': n4_rsp, 'cspec': n4_cspec},
    'b0': {'tte': b0_tte, 'rsp': b0_rsp, 'cspec': b0_cspec}
}
```

```python
# Convert numeric intervals to strings:
background_interval_strings = [f"{start}-{stop}" for start, stop in background_interval]
source_interval_str = f"{source_interval[0]}-{source_interval[1]}"

fluence_plugins = []
time_series = {}

for det in gbm_detectors:
    ts_cspec = TimeSeriesBuilder.from_gbm_cspec_or_ctime(
        det,
        cspec_or_ctime_file=dload[det]["cspec"],
        rsp_file=dload[det]["rsp"],
    )

    ts_cspec.set_background_interval(*background_interval_strings)
    ts_cspec.save_background(f"{det}_bkg.h5", overwrite=True)

    ts_tte = TimeSeriesBuilder.from_gbm_tte(
        det,
        tte_file=dload[det]["tte"],
        rsp_file=dload[det]["rsp"],
        restore_background=f"{det}_bkg.h5",
    )

    time_series[det] = ts_tte
    ts_tte.set_active_time_interval(source_interval_str)
    ts_tte.view_lightcurve(-40, 100)

    fluence_plugin = ts_tte.to_spectrumlike()
    if det.startswith("b"):
        fluence_plugin.set_active_measurements("250-30000")
    else:
        fluence_plugin.set_active_measurements("9-900")

    fluence_plugin.rebin_on_background(1.0)
    fluence_plugins.append(fluence_plugin)
```

```python
from astromodels import PointSource

sbpl_model = SmoothlyBrokenPowerLaw()

sbpl_model.alpha.prior = Truncated_gaussian(lower_bound=-1.5, upper_bound=1, mu=-1, sigma=0.5)
sbpl_model.beta.prior = Truncated_gaussian(lower_bound=-5, upper_bound=-1.6, mu=-2.25, sigma=0.5)
sbpl_model.break_energy.prior = Log_normal(mu=2, sigma=1)
sbpl_model.break_energy.bounds = (None, None)
sbpl_model.K.prior = Log_uniform_prior(lower_bound=1e-3, upper_bound=1e1)
sbpl_model.break_scale.prior = Log_uniform_prior(lower_bound=1e-4, upper_bound=10)


source = PointSource("GRB080916009", ra=0, dec=0, spectral_shape=sbpl_model)
model = Model(source)

data_list = DataList(*fluence_plugins)
jl = JointLikelihood(model, data_list)
```

```python
new_model = clone_model(model)

bayes = BayesianAnalysis(new_model, DataList(*fluence_plugins))

bayes.set_sampler("emcee", share_spectrum=True)
```

```python
bad_bins = []
for i, w in enumerate(n3.bins.widths):
    if w < 5e-2:
        bad_bins.append(i)


edges = [n3.bins.starts[0]]

for i, b in enumerate(n3.bins):
    if i not in bad_bins:
        edges.append(b.stop)

starts = edges[:-1]
stops = edges[1:]


n3.create_time_bins(starts, stops, method="custom")
```

```python
fig = n3.view_lightcurve(use_binner=True)
```

```python
time_resolved_plugins = {}

for k, v in time_series.items():
    v.read_bins(n3)
    time_resolved_plugins[k] = v.to_spectrumlike(from_bins=True)
```

```python
from astromodels import PointSource, Cutoff_powerlaw

# Define the Comptonised spectral model and priors
comp = Cutoff_powerlaw()

comp.K.prior = Log_uniform_prior(lower_bound=1e-5, upper_bound=20)
comp.index.prior = Truncated_gaussian(lower_bound=-3, upper_bound=1, mu=-1, sigma=0.5)
comp.xc.prior = Log_uniform_prior(lower_bound=1.0, upper_bound=5e2)  # 1 to 500 keV, for example
```

```python
base_source = PointSource("GRB080916009", ra=0, dec=0, spectral_shape=comp)
base_model = Model(base_source)

models = []
results = []
analysis = []

for interval in range(12):
    this_model = clone_model(base_model)
    models.append(this_model)

    this_data_list = []
    for k, v in time_resolved_plugins.items():
        pi = v[interval]
        if hasattr(pi, "_rebinner") and pi._rebinner is not None:
            pi.remove_rebinning()
        if k.startswith('b'):
            pi.set_active_measurements("250-30000")
        else:
            pi.set_active_measurements("9-900")
        pi.rebin_on_background(1.0)
        this_data_list.append(pi)

    dlist = DataList(*this_data_list)
    results.append(dlist)
    bayes = BayesianAnalysis(this_model, dlist)
    bayes.set_sampler("emcee", share_spectrum=True)
    bayes.sampler.setup(n_live_points=500, n_iterations=10000)
    bayes.sample()
    analysis.append(bayes)
```

```python
for a in analysis:
    a.restore_median_fit()
    _ = display_spectrum_model_counts(a, min_rate=[20, 20, 20], step=False)
```

```python
fig = plot_spectra(
    *[a.results for a in analysis[::1]],
    flux_unit="erg2/(cm2 s keV)",
    fit_cmap="viridis",
    contour_cmap="viridis",
    contour_style_kwargs=dict(alpha=0.1),
)
```
